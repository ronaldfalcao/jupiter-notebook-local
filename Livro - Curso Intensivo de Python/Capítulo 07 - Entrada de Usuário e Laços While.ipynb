```python
#confirmed_users.py
# Começa com os usuários que precisam ser verificados,
# e com uma lista vazia para armazenar os usuários confirmados

unconfirmed_users = ['alice', 'brian', 'candace']
confirmed_users = []

# Verifica cada usuário até que não haja mais usuários não confirmados
# Transfere cada usuário verificado para a lista de usuários confirmados

while unconfirmed_users:
    current_user = unconfirmed_users.pop()
    
    print("Verifying user: " + current_user.title())
    confirmed_users.append(current_user)
    
# Exibe todos os usuários confirmados...
print("\nThe following users have been confirmed: ")
for confirmed_user in confirmed_users:
    print(confirmed_user.title())
```

```python
#mountain_poll.py

responses = {}

# Define uma flag para indicar que a enquete está ativa
polling_active = True

while polling_active:
    # Pede o nome da pessoa e a resposta
    name = input("\nWhat is your name? ")
    response = input("Which mountain would you like to climb someday? ")
    
    # Armazena a resposta no dicionário
    responses[name] = response
    
    # Descobre se a outra pessoa vai responder à enquete
    repeat = input("Would you like to let another person respond? (yes/no): ")
    if repeat == 'no':
        polling_active = False
        
# A enquete foi concluída, mostra os resultados...
print("\n----------- Poll Results --------------")
for name, response in responses.items():
    print(name + " would like to climb " + response + ".")
```

```python
# O exercício não é claro, ele pede para que usemos a LISTA sandwich_orders e que removamos as incidências
# de 'pastrami'.OK!
# A lista finished_sandwiches não havia sido citada antes, não vi uma forma (sem erro) de remover o item e
# continuar a ler a lista garantindo que ela será percorrida de forma correta.

sandwich_orders = ['pastrami','x-burguer', 'x-salada', 'pastrami', 'americano', 'x-frango', 'pastrami', 'calabresa', 'pastrami']
finished_sandwiches = []

print("Lista inicial de pedidos: ")
print(sandwich_orders)

print("\nInfelizmente o pastrami acabou, todos os pedidos com esse ingrediente serão recusados.")

# removendo as incidências de pastrami...
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')

print("\nLista de pedidos válidos: ")    
print(sandwich_orders)

# processando os pedidos válidos...
while sandwich_orders:
    sandwich_current = sandwich_orders.pop()
    finished_sandwiches.append(sandwich_current)

print("\nLista de pedidos processados: ")    
print(finished_sandwiches)
```

```python
# Define uma flag para indicar que a enquete está ativa
polling_active = True

responses = {}

while polling_active:
    
    name = input("\nPor favor, qual o seu nome? ")
    response = input("Se pudesse ir para um lugar no mundo, para aonde iria? ")
    
    # Armazena a resposta no dicionário
    responses[name] = response
    
    # Descobre se a outra pessoa vai responder à enquete
    repeat = input("Outra pessoa gostaria de responder a nossa enquete? (sim/nao): ")
    if repeat == 'nao':
        polling_active = False
        
# A enquete foi concluída, mostra os resultados...
print("\n----------- Resultados --------------")
for name, response in responses.items():
    print(name + " gostaria de ir para " + response + ".")
```
